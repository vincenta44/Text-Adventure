# Loading the Data

```python
# Load the JSON data with Requests
import json
reports = open("drugs.json")
text = json.load(reports)
```

# Histogram Analysis

For my histogram I will analyze the amount of minors that needed treatment for illicit drug abuse throughout the country. Data throughout one year (2002)

```python
minor_usage = []
states = []

for i in text:
    state = i['State']
    if state not in states:
        minor_usage.append(i['Totals']['Illicit Drugs']['Need Treatment Past Year']['12-17'])
    states.append(state)

import matplotlib.pyplot as plt
plt.hist(minor_usage)
plt.xlabel("Amount of Minors that Recieved Treatment")
plt.ylabel("Number of States")
plt.title("Addiction Treatment in Minors")
```

My histogram shows the amount of minors that recieved treatment (going up by 2000) in 2002 on the x-axis and the number of states that data occurred in is on the y-axis. Each column represents the number of states that had a specific treament count for minors. Most states are within 2000 as seen being the highest bar on the graph. The higher the x-axis goes, the amount of states goes down, which is good. However, some outliers can be found on the right side of the graph which represents a high amount of minors in treatment, but only occurring in few states.

# Trend Analysis

For my secondary analysis I will be still analyzing minors, however I will be looking at the amount of treatment received through time only in Delaware.

```python
minor_usage = []
years = []
for i in text:
    if i['State'] == "Delaware":
        years.append(i['Year'])
        minor_usage.append(i['Totals']['Illicit Drugs']["Need Treatment Past Year"]['12-17'])
    
plt.plot(years, minor_usage)
plt.xlabel("Year")
plt.ylabel("Amount of Minors that Recieved Treatment")
plt.title("Addiction Treatment Rate in Minors in Delaware")
```

This line graph shows the amount of minors who recieved treatment during a specific year. As the years go on on the y-axis, there is subtle change until 2012 where the graph drops off drastically. This means the amount of minors in treatment for illicit abuse has gone down through time which correlates directly to the decreasing amount of minors abusing illicit drugs. 2014 is the lowest point and most recent, according to this data.
